```python
import os 
import glob

# file_names = [
#     "Tdust.csv"
# ]


# condition = True

base_file_dir = "/scratch/dongwooc/scratch_rwa/doga/runs_hden_radius"
run_file_dir = "voronoi_1e5"
# run_file_dir = "trial1"

redshift = "0.0"
```

```python
### Firebox 
galaxy_type = "firebox"

for i in range(11, 1000):
# for i in [11]:
    galaxy_name = f"gal{i}"
    
    file_patterns = [
        f"{galaxy_name}_i0_stats*",
        f"{galaxy_name}_i0_secondary*",
        f"{galaxy_name}_i0_primary*",        
        f"{galaxy_name}_i90_stats*",
        f"{galaxy_name}_i90_secondary*",
        f"{galaxy_name}_i90_primary*",        
        f"{galaxy_name}_i0_sedstats*",
        f"{galaxy_name}_i90_sedstats*",
        f"{galaxy_name}_i0_transparent.fits",
        f"{galaxy_name}_i90_transparent.fits",   
        f"{galaxy_name}_grid_velocity_v.dat",       
        f"{galaxy_name}_properties_opticalMaterialProperties_opticalprops_0.dat",
        f"{galaxy_name}_i0_sedstats.dat",
        f"{galaxy_name}_source_luminosity_luminosities.dat",           
        f"{galaxy_name}_grid_opacity_k.dat"           
    ]
    
    file_dir = f"{base_file_dir}/{galaxy_type}/z{redshift}/{galaxy_name}/{run_file_dir}"
    
    if os.path.exists(file_dir):
        for file_pattern in file_patterns:
            # Use glob to find files matching the pattern
            full_pattern = f"{file_dir}/{file_pattern}"
            matching_files = glob.glob(full_pattern)
            
            for file_path in matching_files:
                try:
                    os.remove(file_path)
                    print(f"{file_path} deleted.")
                except Exception as e:
                    print(f"Error deleting {file_path}:\n{e}")
    else:
        print(f"{file_dir} does not exist.")
```
