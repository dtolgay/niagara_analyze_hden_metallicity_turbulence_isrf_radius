```python
import shutil
import os
```

```python
base_fdir = "/mnt/raid-cita/dtolgay/FIRE/post_processing_fire_outputs/skirt/runs_hden_radius"
redshift = "0.0"


files_to_copy = [
    "skirt_gas.txt",
    "skirt_star.txt",
    "comprehensive_gas.txt",
    "comprehensive_star.txt"
]
```

```python
galaxy_type = "firebox"

for file_number in range(0,int(1e3)):
        
    print("\n\n")
        
    try:
        
        path_to_folder_to_copy = f"{base_fdir}/{galaxy_type}/z{redshift}/gal{file_number}/voronoi_1e6"
        destination_file_path = f"{base_fdir}/{galaxy_type}/z{redshift}/gal{file_number}/voronoi_1e5"
        
        # Check if paste folder exists. If not create. 
        if os.path.isdir(destination_file_path):
            print(f"The folder '{destination_file_path}' exists.")
        else:
            print(f"The folder '{destination_file_path}' does not exist. Creating folder.")
            os.makedirs(destination_file_path)
        
        # Copy files into paste folders
        for file in files_to_copy:
            
            path_to_file = f"{path_to_folder_to_copy}/{file}" 
        
            # Copy the file with it's new name
            shutil.copy(path_to_file, destination_file_path)
        
            print(f"{path_to_file} -> {destination_file_path}")
        
    except IOError as e:
        print(f"Unable to copy file. {e}")

    except PermissionError as e:
        print(f"Permission denied. {e}")    
```
