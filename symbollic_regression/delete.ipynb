```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x1 = np.logspace(0.1, 10, 100)
x2 = np.logspace(1, 100, 100)
x3 = np.logspace(2, 500, 100)

a, b, c, d = 1.5, 2.5, 3.5, 4.5

log_fx = a * np.log10(x1) + b * np.log10(x2) + c * np.log10(x3) + d
fx = x1**a * x2**b * x3**c * 10**d 

plt.scatter(
    log_fx, np.log10(fx), c='blue', s=10, label='Data Points'
)
```
