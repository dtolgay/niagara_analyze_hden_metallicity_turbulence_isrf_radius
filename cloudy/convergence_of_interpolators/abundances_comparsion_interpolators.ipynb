```python
import pandas as pd 
import numpy as np 

from tools_tolgay import functions_readfiles as readfiles

import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams['mathtext.fontset'] = 'cm'
rcParams['font.family'] = 'serif'
rcParams['font.size'] = 16
rcParams['figure.figsize'] = (12, 6)
rcParams['figure.dpi'] = 100
```

```python
base_fdir = "/scratch/dtolgay/cloudy_runs/z_0/m12i_res7100_md_test"

runs_m12i_res7100_md = {
    "RBFInterpolator_N100": {
        "label": "N100",
        "file_name": "abundance_RBFInterpolator_N100_smoothingLength_interpolation.txt",
        "color": "purple",
    },
    "RBFInterpolator_N200": {
        "label": "N200",
        "file_name": "abundance_RBFInterpolator_N200_smoothingLength_interpolation.txt",
        "color": "magenta",
    },
    "RBFInterpolator_N300": {
        "label": "N300",
        "file_name": "abundance_RBFInterpolator_N300_smoothingLength_interpolation.txt",
        "color": "cyan",
    },
    "RBFInterpolator_N1000": {
        "label": "N1000",
        "file_name": "abundance_RBFInterpolator_N1000_smoothingLength_interpolation.txt",
        "color": "blue",
    },
    "expected": {
        "label": "expected",
        "file_name": "abundance_smoothingLength_expected.txt",
        "color": "brown",
    },
}

for key, value in runs_m12i_res7100_md.items():
    path_to_file = f"{base_fdir}/{value['file_name']}"
    print(path_to_file)
    data, _ = readfiles.read_interpolated_files_usingFilePath2(path = path_to_file, interpolation_type = "abundance")
    data['Mco'] = data['mass'] * data['fCO']
    data['Mh2'] = data['mass'] * data['fh2']
    value['data'] = data
    



# Plot the distribution of luminosities for different interpolators

line = "Mco"
bins = np.linspace(-5, 8, 100)
for run in runs_m12i_res7100_md.keys():

    total_luminosity = np.sum(runs_m12i_res7100_md[run]["data"][line])

    luminosities = runs_m12i_res7100_md[run]["data"][line]
    plt.hist(
        np.log10(luminosities), 
        bins = 30, alpha = 0.5, 
        label = f"run: {runs_m12i_res7100_md[run]['label']}, total Mco = {total_luminosity:.2e} K km/s pc^2", 
        color = runs_m12i_res7100_md[run]["color"]
    )
    plt.axvline(np.median(np.log10(luminosities)), color = runs_m12i_res7100_md[run]["color"], linestyle = "--")


plt.xlabel(f"{line} [Msolar]")
plt.ylabel("Number of gas particles")
plt.title("m12i res7100 md test")
plt.legend()

figure_name = f"{line}_m12i_res7100_md_test.png"
plt.savefig(figure_name)
print(f"Figure saved as {figure_name}")
plt.show()



line = "Mh2"
bins = np.linspace(-5, 8, 100)
for run in runs_m12i_res7100_md.keys():

    total_luminosity = np.sum(runs_m12i_res7100_md[run]["data"][line])

    luminosities = runs_m12i_res7100_md[run]["data"][line]
    plt.hist(
        np.log10(luminosities), 
        bins = 30, alpha = 0.5, 
        label = f"run: {runs_m12i_res7100_md[run]['label']}, total Mh2 = {total_luminosity:.2e} K km/s pc^2", 
        color = runs_m12i_res7100_md[run]["color"]
    )
    plt.axvline(np.median(np.log10(luminosities)), color = runs_m12i_res7100_md[run]["color"], linestyle = "--")


plt.xlabel(f"{line} [Msolar]")
plt.ylabel("Number of gas particles")
plt.title("m12i res7100 md test")
plt.legend()

figure_name = f"{line}_m12i_res7100_md_test.png"
plt.savefig(figure_name)
print(f"Figure saved as {figure_name}")
plt.show()

############################

plt.scatter(
    np.log10(runs_m12i_res7100_md["expected"]["data"]["Mh2"]),
    np.log10(runs_m12i_res7100_md["RBFInterpolator_N200"]["data"]["Mh2"]), 
    label = "expected", color = "black"
)
plt.plot([ -5, 8], [-5, 8], linestyle = "--", color = "red")
plt.xlabel("expected Mh2 [Msolar]")
plt.ylabel("RBFInterpolator_N200 Mh2 [Msolar]")
```
