```python
import os
import shutil

base_file_name = "temperature_regularGridInterpolator_linear_smoothingLength"

base_file_dir = "/home/m/murray/dtolgay/scratch/post_processing_fire_outputs/skirt/runs_hden_radius"

file_name = f"{base_file_name}.txt"
destination_file_dir = f"/home/m/murray/dtolgay/scratch/post_processing_fire_outputs/skirt/migrated_files_from_cita/{base_file_name}"


redshift = "0.0"
run_file_dir = "voronoi_1e6"
```

```python
galaxy_type = "firebox"

for galaxy_number in range (int(1e3)):
# for galaxy_number in [0]:
    
    galaxy_name = f"gal{galaxy_number}"
    file_dir = f"{base_file_dir}/{galaxy_type}/z{redshift}/{galaxy_name}/{run_file_dir}"

    try:  
        # Check if file exits
        file_path = f"{file_dir}/{file_name}"

        if os.path.exists(file_path):
            destination_file_path = f"{destination_file_dir}/{galaxy_name}.txt"
            # Copy the file with it's new name
            shutil.copy(file_path, destination_file_path)            
            
            print(f"{file_path} ---> {destination_file_path}")
        
        else: 
            print(f"{file_path} does not exist...")
        
    except Exception as e:
        print(f"Error occured:\n{e}")
        
    print("\n")
```
